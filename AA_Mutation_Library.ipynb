```python
########### Library ##############
import pandas as pd
import numpy as np
import time

########### Run ############## Positioning Method
seq = input('서열을 집어 넣어 주세요! \n')
position = [100, 101, 102, 103, 142, 143, 168, 185, 186, 187, 190, 211, 215, 216, 217, 218, 219,221,222,515,539,541]
position = [i-1 for i in position] ## 서열이 1번 부터 시작하는 경우
print([seq[int(q)] for q in position])
time.sleep(3)


print(r'############## Procceed ##############')
AA_list = ['G', 'A','V','C','P','L','I','M','W','F','K','R','H','S', 'T', 'Y', 'N', 'Q', 'D', 'E']
candidates = [ [seq[:i] + AA + seq[i+1:], seq[i]+str(i+1)+AA] for AA in AA_list for i in position]
if len(seq.split(' ')) >= 2 : 
    raise('Error 1 : 서열 내 공백 존재')

########### save a csv file for Kcat analysis ##############
print(r'[1] Saving seq lists to csv file')
pd.DataFrame(candidates, columns  = [['Seq','Mutation']]).to_csv('seq_mutation_candidates_Kcat.csv', index = False, header = None)
print(r' --> Success')
```

```python
########### Library ##############
import pandas as pd
import numpy as np

########### Run ##############
seq = input('서열을 집어 넣어 주세요! \n')
print(r'############## Procceed ##############')
AA_list = ['G', 'A','V','C','P','L','I','M','W','F','K','R','H','S', 'T', 'Y', 'N', 'Q', 'D', 'E']
candidates = [seq[:i] + AA + seq[i+1:] for AA in AA_list for i in range(len(seq))]
if len(candidates[0].split(' ')) >= 2 : 
    raise('Error 1 : 서열 내 공백 존재')
    
print('서열길이 : ', len(seq), ' /생성된 데이터 개수 : ', len(np.unique(np.array(candidates))))

########### save a csv file for Kcat analysis ##############
print(r'[1] Saving seq lists to csv file')
pd.DataFrame(np.unique(np.array(candidates))).to_csv('seq_mutation_candidates_Kcat.csv', index = False, header = None)
print(r' --> Success')

########### save a csv file for Km analysis ##############
print(r'[2] Saving seq lists to fasta file')
fasta = open("results.fasta" , 'w')
for num,seqs in enumerate(np.unique(np.array(candidates))):
    if num == 0 :
        fasta.write('>Seq' + str(num) +'_test\n' + seqs )
    else :
        fasta.write('\n>' + 'Seq' + str(num) +'_test\n' + seqs )    
fasta.close()
print(r' --> Success', '\n############## Finished ##############')

########### Original seq index searching ##############
data = pd.DataFrame(np.unique(np.array(candidates)))
print('Original Seq index : ', data.loc[data[0]==seq,:].index[0])
print('', data.loc[data[0]==seq,:][0].tolist()[0])

del candidates

#### ETC ####
# candidates = [seq[:i] + AA + seq[i+1:] for AA in AA_list for i in range(len(seq))]
# [[seq[:i] + AA + seq[i+1:] for AA in AA_list] for i in range(len(seq))]
```
