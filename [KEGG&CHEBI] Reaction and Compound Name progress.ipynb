```python
def printProgressPercent(count, Data):
    from IPython.display import clear_output
    clear_output(wait=True)
    print( str( round(100*(count+1)/len(Data),2 )) +'% Processing',
      end='\r',
      flush=True)
    return
```

```python
file_path = r'chebi_compound_Name_v2.tsv'
data = pd.read_csv(file_path,
                  sep = '\t',
                   header=None)

file_path = r'rhea_reaction'
data_2 = pd.read_csv(file_path,
                  sep = '\t',
                   header=None)
data_2.head()

reactionData_matched = data_2.loc[data_2[2] != 'NoMatched' ]
reactionData_matched_cleen = reactionData_matched.dropna().reset_index(drop = True)
display(reactionData_matched_cleen.head())



import re
import numpy as np

# 숫자와 글자가 아닌 문자 추출
lst = [''.join(re.findall(r'[^A-Za-z0-9]+', input_string)) for input_string in data[1]]
print('Before :', np.unique(lst))

## Compound 이름에서 기호들 다 제거 ## 문자 숫자 공백만 있음
data[1] = data[1].str.lstrip() # 앞 공백 제거 #
data[1] = data[1].str.replace('(\+)', 'plus').str.replace('(\-)', 'minus')
data[1] = data[1].str.replace('(', '').str.replace(')', '').str.replace('+', '').str.replace('-', '').str.replace(r'{', '').str.replace(r'}', '')
data[1] = data[1].str.replace(',', '').str.replace('[', '').str.replace(']', '').str.replace(r"'", '').str.replace(r".", '').str.replace(r">", '')
data[1] = data[1].str.replace('<', '').str.replace(r'/', '').str.replace(r':', '')

lst = [''.join(re.findall(r'[^A-Za-z0-9]+', input_string)) for input_string in data[1]]
print('After :', np.unique(lst))

data.head()
```
